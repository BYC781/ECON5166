```python
import pandas as pd
import numpy as np
data = pd.read_parquet('NYSE_trade(homework).parquet', engine='pyarrow')

data['lagprice1'] = data['PRICE'].shift(1)
data['lagprice2'] = data['PRICE'].shift(2)
data['lagprice3'] = data['PRICE'].shift(3)
data['lagprice4'] = data['PRICE'].shift(4)
data['r_t'] = (data["PRICE"]-data["lagprice1"])/data["lagprice1"]
data['r_t_1'] = (data["lagprice1"]-data["lagprice2"])/data["lagprice2"]
data['r_t_2'] = (data["lagprice2"]-data["lagprice3"])/data["lagprice3"]
data['r_t_3'] = (data["lagprice3"]-data["lagprice4"])/data["lagprice4"]
```

```python
from sklearn.preprocessing import StandardScaler

trainX = [D1_train_X, D2_train_X, D3_train_X, D4_train_X, D5_train_X]
trainY = [D1_train_Y, D2_train_Y, D3_train_Y, D4_train_Y, D5_train_Y]
testX = [D1_test_X, D2_test_X, D3_test_X, D4_test_X, D5_test_X]
testY = [D1_test_Y, D2_test_Y, D3_test_Y, D4_test_Y, D5_test_Y]

for k in range(5):
  scaler = StandardScaler()
  trainX[k] = scaler.fit_transform(trainX[k])
  trainY[k] = scaler.fit_transform(trainY[k])
  testX[k] = scaler.fit_transform(testX[k])
  testY[k] = scaler.fit_transform(testY[k])

```

```python
from sklearn.linear_model import Lasso

alpha = np.linspace(1, 10, num=10)
tot_sse =[]

for i in range(10):
  sse = 0
  for k in range(5):
    model = Lasso(alpha=alpha[i], max_iter=10000)
    model.fit(trainX[k],trainY[k])
    pred = model.predict(testX[k])
    sse += (np.sum((testY[k].flatten() - pred) ** 2))
  tot_sse.append(sse)

print(tot_sse)
print('best alpha:',alpha[(np.argmin(tot_sse))],' minimum sum of squared error:',min(tot_sse))
```
